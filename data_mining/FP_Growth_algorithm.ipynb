```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns.fpgrowth import fpgrowth
from mlxtend.frequent_patterns import association_rules
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

# **Step1: Load the file text and convert as list of Transactions**

```python
def Load_data(filename):
    with open(filename, "r") as infile:
            transactions = [set(line.rstrip("\n").split(";"))
                            for line in infile]

    return transactions

f = '/content/drive/MyDrive/Colab Notebooks/categories.txt'
transactions= Load_data(f)

#print the category text file
#transactions
```

# **Step2: Generate an encoded dataframe**

```python
# instantiate a transaction encoder
transcoder = TransactionEncoder()

# fit the transaction encoder using the list of transaction tuples
transcoder.fit(transactions)

# transform the list of transaction tuples into an array of encoded transactions
transarray = transcoder.transform(transactions)

# convert the array of encoded transactions into a dataframe
trans_df = pd.DataFrame(transarray, columns = transcoder.columns_)
trans_df
```

# **Step3: Compute the frequent itemsets using fpgriowth from mlxtend**

```python
min_support = 0.01
frequent_itemsets = fpgrowth(trans_df, min_support=min_support, use_colnames = True)

# print the frequent itemsets
frequent_itemsets
```

**Save the Frequent Category itemsets including their absolute supports into text file**

```python
pattren = frequent_itemsets.to_csv('/content/drive/MyDrive/pattern.txt', sep='\t')
```

# **Step4: Compute the association rules based on the frequent itemsets**

```python
# compute and print the association rules
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.01)
```

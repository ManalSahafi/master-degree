```python
def remove_punctuations(text):
    no_punc="".join([x for x in text if x not in string.punctuation])
    return no_punc
```

```python
df['remove_punctuations']= df['Text'].apply(lambda x: remove_punctuations(x))
df.head()
```

```python
df['Tokens']= df.apply(lambda row: nltk.word_tokenize(row['remove_punctuations']), axis=1)
df.head()
```

```python
def remove_stopword(text):
    words = [w for w in text if w not in stopwords.words('english')]
    return words
```

```python
df['remove_stopword']= df['Tokens'].apply(lambda x: remove_stopword(x))
df.head()
```

```python
# Init the Wordnet Lemmatizer
lemmatizer = WordNetLemmatizer()

#define function for lemmatization
def lemmatize_text(text):
    return [lemmatizer.lemmatize(w) for w in text]

df['lemmatized_words']= df['remove_stopword'].apply(lambda x: lemmatize_text(x))

df.head()
```

```python
pos = PorterStemmer()
def stemming(text):
    return [pos.stem(w) for w in text]

df['stemmtized_words']= df['remove_stopword'].apply(lambda x: stemming(x))

df.head()
```

```python
#if we want use n_grams model without tokinaztion we need to remove stopwords first
def generate_N_grams(text,ngram=1):
    words=[word for word in text.split(" ") if word not in set(stopwords.words('english'))]  
    print("Sentence after removing stopwords:",words)
    temp=zip(*[words[i:] for i in range(0,ngram)])
    ans=[' '.join(ngram) for ngram in temp]
    return ans
```

```python
generate_N_grams("The sun rises in the east",2)
```
